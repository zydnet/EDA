```python
import pandas as pd
df=pd.read_csv('student_ml_dataset.csv')
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
print(df.head())
print(df.info())
```

```python
import matplotlib.pyplot as plt
df.hist(figsize = (10,10),bins=50, color='red', alpha=0.75, rwidth=0.9)
plt.title('Histogram', fontsize=15, fontweight='bold')
plt.xlabel('Values', fontsize=10, fontweight='bold')
plt.ylabel('Frequency', fontsize=10, fontweight='bold')
plt.show()

df['total_score'] = df['Math_Marks'] + df['Physics_Marks'] + df['Chemistry_Marks']

plt.hist(df['total_score'], bins=10, color='red',edgecolor='black',linewidth=1.2)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('THE TOTAL MARKS OF THE SUBJECTS ARE')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
s=pd.read_csv('student_ml_dataset.csv')
print(s.describe())

plt.scatter(s['Age'],s['Math_Marks'],color='red', alpha=0.75)
plt.scatter(s['Age'],s['Chemistry_Marks'])
plt.scatter(s['Age'],s['Physics_Marks'])

plt.xlabel('age')
plt.ylabel('marks')
plt.show()

plt.figure(figsize=(5,6))
plt.hist(s['Chemistry_Marks'],bins=10,edgecolor='black',alpha=0.5,color='r')
plt.title('file')
plt.xlabel('part time')
plt.ylabel('freq')

plt.show()


```
